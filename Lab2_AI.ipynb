```python
import skfuzzy
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = np.arange(0, 1.0001, 0.0001)
```

```python
#task 1.1
triangle_f = skfuzzy.trimf(x, [0.1, 0.3, 0.6])
sns.lineplot(x=x, y=triangle_f)
```

```python
#task 1.2
trap_f = skfuzzy.trapmf(x, [0.1, 0.3, 0.4, 0.6])
sns.lineplot(x=x, y=trap_f)
```

```python
#task 2.1
gauss_f = skfuzzy.gaussmf(x, 0.4, 0.1)
sns.lineplot(x=x, y=gauss_f)
```

```python
#task 2.2
gauss2_f = skfuzzy.gauss2mf(x, 0.4, 0.1, 0.4, 0.2)
sns.lineplot(x=x, y=gauss2_f)
```

```python
#task 3
j = np.arange(0, 10, 0.1)
bell_f = skfuzzy.gbellmf(j, 2, 2, 6)
sns.lineplot(x=j, y=bell_f)
```

```python
#task 4.1
sig_f = skfuzzy.sigmf(x, 0.6, 20)
sns.lineplot(x=x, y=sig_f)
```

```python
#task 4.2
d_sig_f = skfuzzy.psigmf(x, 0.3, 15, 0.5, 15)
sns.lineplot(x=x, y=d_sig_f)
```

```python
#task 4.3
d_u_sig_f = skfuzzy.dsigmf(x, 0.2, 15, 0.7, 30)
sns.lineplot(x=x, y=d_u_sig_f)
```

```python
#task 5.1
pi_f = skfuzzy.pimf(x, 0.1, 0.3, 0.5, 0.9)
sns.lineplot(x=x, y=pi_f)
```

```python
#task 5.2
s_f = skfuzzy.smf(x, 0.3, 0.8)
sns.lineplot(x=x, y=s_f)
```

```python
#task 5.3
z_f = skfuzzy.zmf(x, 0.2, 0.8)
sns.lineplot(x=x, y=z_f)
```

```python
#task 6
df = pd.DataFrame()
df["x"] = x
df["g1"] = skfuzzy.gaussmf(x, 0.2, 0.2)
df["g2"] = skfuzzy.gaussmf(x, 0.8, 0.2)
_, df["and"] = skfuzzy.fuzzy_and(x, df["g1"].to_numpy(), x, df["g2"].to_numpy())
_, df["or"] = skfuzzy.fuzzy_or(x, df["g1"].to_numpy(), x, df["g2"].to_numpy())
```

```python
melted = pd.melt(df[["x", "and", "or"]], id_vars=['x'])
sns.lineplot(data=melted, x='x', y='value', hue='variable')
```

```python
#task 7
df = pd.DataFrame()
df["x"] = x
df["g1"] = skfuzzy.gaussmf(x, 0.2, 0.2)
df["g2"] = skfuzzy.gaussmf(x, 0.8, 0.2)
df["and"] = df["g1"] * df["g2"]
df["or"] = df["g1"] + df["g2"] - df["g1"] * df["g2"]
```

```python
melted = pd.melt(df, id_vars=['x'])
sns.lineplot(data=melted, x='x', y='value', hue='variable')
```

```python
#task 8
df = pd.DataFrame()
df["x"] = x
df["g"] = skfuzzy.gaussmf(x, 0.3, 0.1)
df["!g"] = skfuzzy.fuzzy_not(df["g"])
melted = pd.melt(df, id_vars=['x'])
sns.lineplot(data=melted, x='x', y='value', hue='variable')
```
